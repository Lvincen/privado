```python
for i, elemento in enumerate(wall):
    R = np.array(elemento['R'], dtype=np.float64)
    Z = np.array(elemento['Z'], dtype=np.float64)

    # Cerrar contorno si es necesario
    if R[0] != R[-1] or Z[0] != Z[-1]:
        R = np.append(R, R[0])
        Z = np.append(Z, Z[0])
    
    nombre = f'Componente {i+1}'  # Asignación automática
    plt.plot(R, Z, label=nombre)

plt.xlabel('R (m)')
plt.ylabel('Z (m)')
plt.title('Contornos desde archivo pickle')
plt.axis('equal')
plt.legend()
plt.grid(True)
plt.show()
```
